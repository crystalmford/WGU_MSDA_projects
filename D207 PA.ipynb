A1: Question for Analysis 

For this project, I am using the same churn dataset as I did in previous projects. I'm interested in determining whether the monthly charges had an effect on the churn rate of customers. The churn rate in question is the percentage rate at which customers stop subscribing, or cancel subscriptions to the service. My question is "Are customers paying a higher monthly charge more likely to have higher churn rates?" The null hypothesis would be that a higher monthly charge has no significant effect on the churn rate of customers.

H0 (Null Hypothesis):  
A customer's monthly charge has no effect on the churn rate of customers.

HA (Alternvative Hypothesis): 
Higher monthly charges will have higher churn rates.

A2: Benefit from Analysis 

Stakeholders will find benefit from this research because in business it is always important to look for ways to increase profit. If it turns out that having a higher monthly charge results in a higher churn rate, it would be advisable to lower prices in some fashion, either with temporary discounts and promotions, or lower prices across the board. According to the scenario consideration sheet, it costs 10 times more to acquire a new customer than to retain an existing one. This makes it incredibly important to figure out ways to keep customers and keep them happy, while also making profits. 


A3: Data Identification 

In order to answer the research question the following variables are necessary: 

Churn (categorical) - This column indicates whether the customer has discontinued service within the last month. (yes/no) 

MonthlyCharge (quantitative) This column indicates the monthly charge of the customers, as an average. (continuous)

```python
#Import dataframe and prep for data analysis
#import churn_clean pdf as df
import pandas as pd
df = pd.read_csv(r"C:\Users\Crystal\OneDrive\Desktop\fordversity\d207\churn_clean.csv")
df.head()

df.info()
print()

#Convert Churn to categorical datatype from object
df['Churn'] = df.Churn.astype('category') 

#Conduct two sample t-test on churn and monthly charge variables
import scipy.stats as stats


#Create two arrays
c_yes= df[df.Churn == 'Yes'].MonthlyCharge
c_no= df[df.Churn == 'No'].MonthlyCharge

#Perform the two-sample t-test
t_result= stats.ttest_ind(c_yes, c_no)
print(t_result)

p_value= t_result[1]
print(p_value)

#Test significance
alpha= 0.05
if (t_result[1] < alpha):
    print('Customers with churn have higher monthly charges than those without churn')
else: print('No significant difference found')
    

```

B3: Justification
For this analysis, I was looking to compare whether the means of monthly charges for customers with a yes in the churn category are equal to the means of the monthly charges for customers with a no in the churn category. Or in simple english, are customers who have recently churned paying higher monthly charges than customers who have not? I chose to use a two sample t-test because that type of test is used to determine whether the means of two populations are equal. I had two populations to test with the Churn variable - yes or no. Using this t-test told me that the average monthly charge of people who have recently churned is not equal to the average monthly charge of people who have not recently churned.

E1: Results of Analysis
After performing a two sample t-test, I found that the monthly charges of customers who have recently churned are in fact higher than the monthly charges for customers who have not. The p-value returned was far less than the alpha of 0.05 which means that the null hypothesis should probably be rejected. The smaller the p-value is the less likely it is that these results occurred by random chance, which means it is more likely that the two variables are related.
